```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([2,3,4,5,6,7,8,9])
plt.scatter(x,y)
```

```python
def linear_regi(x,y):
    m=0
    c=0
    lr=0.001
    for i in range(100):
        m=m-lr*(np.sum((m*x+c-y)*x))
        c=c-lr*(np.sum(m*x+c-y))
    return m,c #optimal values
(m,c)=linear_regi(x,y)
plt.scatter(x,y)
plt.plot(x,m*x+c)
plt.show()
        
```
